```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ficheros=[]
datos=[]
```

```python
with open("software/salidas/localSearchEvol.txt",'r') as f:
    lines= f.readlines()

lines=lines[1:]

i=0
while i < len(lines):
    ficheros.append(lines[i].replace('data/','').replace('\n',''))
    datos.append(list(map(float,lines[i+1].split(', ')[:-1])))
    i+=2
```

```python
for i in range(len(ficheros)):
    name=ficheros[i]
    y = datos[i]
    x = list(range(len(y)))
    plt.plot(x,y,'b-')
    plt.title(name)
    #plt.yticks(np.arange(min(y), max(y), 100))
    plt.xlabel('Solución')
    plt.ylabel('Diversidad')
    #plt.savefig('memoria/imgs/LS2E'+name[:-4].replace('_','-')+'.png')
    plt.show()
```
