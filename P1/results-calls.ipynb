```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ficheros=[]
datos=[]
```

```python
with open("software/salidas/localSearch2.txt",'r') as f:
    lines= f.readlines()

lines=lines[1:]

i=0
while i < len(lines):
    ficheros.append(lines[i].replace('data/','').replace('\n',''))
    datos.append(int(list(map(float,lines[i+1].split('\t')))[2]))
    i+=2
```

```python
datos
```

```python
df = pd.DataFrame({'File':[],'Calls':[]})
```

```python
for i in range(len(ficheros)):
    new_row={'File':ficheros[i][:-4],'Calls':datos[i]}
    df=df.append(new_row, ignore_index=True)
```

```python
df.index=df.File
df.drop(columns=['File'],inplace=True)
```
